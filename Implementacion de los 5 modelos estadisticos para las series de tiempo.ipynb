```python
def group(data, column_name, frec):

        """
        Esta funcion agrupa por serie de tiempo en dias para obtener la media los datos
        """
        data = data.groupby([pd.Grouper(key=column_name, freq=frec)]).mean() 
        data = data.reset_index()
        data[column_name] = pd.to_datetime(data[column_name]) 
        data[column_name] = data[column_name].sort_values(ascending=False)
        data = data.set_index(column_name)

        return data
```

```python
def RMSE_score(y_true, y_pred):
    
    y1 = mean_squared_error(y_true[['Precipitacion']].iloc[-7], y_pred[['predicted_mean']].iloc[-7])
    y2 = mean_squared_error(y_true[['Precipitacion']].iloc[-6], y_pred[['predicted_mean']].iloc[-6])
    y3 = mean_squared_error(y_true[['Precipitacion']].iloc[-5], y_pred[['predicted_mean']].iloc[-5])
    y4 = mean_squared_error(y_true[['Precipitacion']].iloc[-4], y_pred[['predicted_mean']].iloc[-4])
    y5 = mean_squared_error(y_true[['Precipitacion']].iloc[-3], y_pred[['predicted_mean']].iloc[-3])
    y6 = mean_squared_error(y_true[['Precipitacion']].iloc[-2], y_pred[['predicted_mean']].iloc[-2])
    y7 = mean_squared_error(y_true[['Precipitacion']].iloc[-1], y_pred[['predicted_mean']].iloc[-1])

    mean = sqrt((y1 + y2 + y3 + y4 + y5 + y6 + y7)/7 ) 
    return mean 
```

```python
def MSE_score(y_true, y_pred):
    
    y1 = mean_squared_error(y_true[['Precipitacion']].iloc[-7], y_pred[['predicted_mean']].iloc[-7])
    y2 = mean_squared_error(y_true[['Precipitacion']].iloc[-6], y_pred[['predicted_mean']].iloc[-6])
    y3 = mean_squared_error(y_true[['Precipitacion']].iloc[-5], y_pred[['predicted_mean']].iloc[-5])
    y4 = mean_squared_error(y_true[['Precipitacion']].iloc[-4], y_pred[['predicted_mean']].iloc[-4])
    y5 = mean_squared_error(y_true[['Precipitacion']].iloc[-3], y_pred[['predicted_mean']].iloc[-3])
    y6 = mean_squared_error(y_true[['Precipitacion']].iloc[-2], y_pred[['predicted_mean']].iloc[-2])
    y7 = mean_squared_error(y_true[['Precipitacion']].iloc[-1], y_pred[['predicted_mean']].iloc[-1])

    mean = (y1 + y2 + y3 + y4 + y5 + y6 + y7)/7 
    return mean 
```

```python
def group(data, column_name, frec):

        """
        Esta funcion agrupa por serie de tiempo en dias para obtener la media los datos
        """
        data = data.groupby([pd.Grouper(key=column_name, freq=frec)]).mean() 
        data = data.reset_index()
        data[column_name] = pd.to_datetime(data[column_name]) 
        data[column_name] = data[column_name].sort_values(ascending=False)
        data = data.set_index(column_name)

        return data
```
